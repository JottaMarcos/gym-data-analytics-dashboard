```
import pandas as pd

# 1. Carregar a base com o separador correto (ponto e vírgula)
# Isso vai fazer o Python separar as colunas direitinho
df = pd.read_csv('Base_Academia.csv', encoding='latin1', sep=';')

# Limpa qualquer espaço invisível que possa ter nos nomes das colunas
df.columns = df.columns.str.strip()

# 2. Criar a inteligência de Perfis (Fanático, Frequentador, Turista)
def definir_perfil(f):
    try:
        val = float(f)
        if val >= 4: return 'Fanático'
        elif val >= 2: return 'Frequentador'
        else: return 'Turista'
    except: return 'Turista'

df['Perfil_Python'] = df['Frequencia_Semanal'].apply(definir_perfil)

# 3. Criar a métrica de Churn (Baseado no Status)
# Se o status for Cancelado ou Inativo, o churn é 1
status_saida = ['cancelado', 'inativo', 'atrasado']
df['Churn_Python'] = df['Status'].apply(lambda x: 1 if str(x).lower().strip() in status_saida else 0)

# 4. Cálculo de LTV Estimado (Regra de Negócio)
# Multiplicamos a mensalidade pelo tempo que achamos que cada perfil fica
multiplicador = {'Fanático': 12, 'Frequentador': 6, 'Turista': 3}
df['LTV_Estimado'] = df.apply(lambda row: row['Valor_Mensalidade'] * multiplicador[row['Perfil_Python']], axis=1)

# 5. Salvar a base pronta para o GitHub
df.to_csv('Base_Academia_Tratada.csv', index=False, encoding='utf-8-sig')

print("✅ SUCESSO! Base tratada e colunas criadas.")
print(f"Colunas agora: {df.columns.tolist()}")
df[['ID_Transacao', 'Perfil_Python', 'LTV_Estimado', 'Churn_Python']].head()
```
